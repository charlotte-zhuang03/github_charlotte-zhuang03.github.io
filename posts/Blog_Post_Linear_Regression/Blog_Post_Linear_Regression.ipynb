### Creating and Plotting Data

```python
import numpy as np
from matplotlib import pyplot as plt

def pad(X):
    return np.append(X, np.ones((X.shape[0], 1)), 1)

def LR_data(n_train = 100, n_val = 100, p_features = 1, noise = .1, w = None):
    if w is None: 
        w = np.random.rand(p_features + 1) + .2
    
    X_train = np.random.rand(n_train, p_features)
    y_train = pad(X_train)@w + noise*np.random.randn(n_train)

    X_val = np.random.rand(n_val, p_features)
    y_val = pad(X_val)@w + noise*np.random.randn(n_val)
    
    return X_train, y_train, X_val, y_val
```

```python
n_train = 100
n_val = 100
p_features = 1
noise = 0.2

# create some data
X_train, y_train, X_val, y_val = LR_data(n_train, n_val, p_features, noise)

# plot it
fig, axarr = plt.subplots(1, 2, sharex = True, sharey = True)
axarr[0].scatter(X_train, y_train)
axarr[1].scatter(X_val, y_val)
labs = axarr[0].set(title = "Training", xlabel = "x", ylabel = "y")
labs = axarr[1].set(title = "Validation", xlabel = "x")
plt.tight_layout()
```

### Part I: Implementing Linear Regression in Two Ways

#### Method 1: Using analytical formula

In the lecture notes, we can find an analytical formula that can calculate the optimal weight vector $\hat{w}$:

$$\hat{w} = (X^{T}X)^{-1}X^{T}y$$

Therefore, we can use the formula w_hat = np.linalg.inv(X.T@X)@X.T@y to compute the optimal weight vector $\hat{w}$ directly. The method is denoted as fit_analytical(X, y) in linear_regression.py

##### Application of Method 1

```python
from linear_regression import LinearRegression
LR = LinearRegression()
LR.fit_analytical(X_train, y_train) # I used the analytical formula as my default fit method

print(f"Training score = {LR.score(X_train, y_train).round(4)}")
print(f"Validation score = {LR.score(X_val, y_val).round(4)}")
```

##### Get optimal weight vector

```python
LR.w
```

#### Method 2: Using gradient descent method

This method basically follows the gradient descent method as other blog posts. 

The formula of the gradient is:
$$\nabla L(w) = 2X^{T}(Xw-y)$$

For every step in the updating process:

We subtract $\alpha \nabla L(w^{(t)})$ from the current weight vector $ w^{(t)}$ to update the weight vector w until convergence.

This method is denoted as fit_gradient(X, y) in linear_regression.py

##### Application of Method 2:

```python
LR2 = LinearRegression()

LR2.fit_gradient(X_train, y_train, alpha = 0.01, max_iter = 1e2)
LR2.w
```

```python
plt.plot(LR2.score_history)
labels = plt.gca().set(xlabel = "Iteration", ylabel = "Score")
```

### Part II: Experiments

This experiment requires to increase p_features, the number of features used, while holding n_train, the number of training points, constant. 

In this experiment, p_features will be increased all the way to n_train - 1. A plot will be made to illustrate the change of scores corresponding to the increase of p_features.

```python
n_train = 100
n_val = 100
noise = 0.2

training_score_history = []
validation_score_history = []
p_features = 1
for i in range (n_train-1):
    X_train, y_train, X_val, y_val = LR_data(n_train, n_val, p_features, noise)
    LR.fit_analytical(X_train, y_train)
    training_score_history.append(LR.score(X_train, y_train).round(4))
    validation_score_history.append(LR.score(X_val, y_val).round(4))
    p_features += 1
plt.plot(np.arange(n_train-1), training_score_history, label = "Training Score")
plt.plot(np.arange(n_train-1), validation_score_history, label = "Validation Score")
labels = plt.gca().set(xlabel = "p_features", ylabel = "Scores")
legend = plt.legend()
```

##### Observation: 

The training score does not change much, but the validation score experienced a huge drop and rise (large-scale fluctuation) when p_features is large. This might be due to the model is overfitting to the training dataset, therefore it is not able to predict the validation set very well. This causes a stark decrease in the validation score.

### Part III: LASSO Regularization

In this part, I will use the LASSO algorithm instead of the regular linear regression algorithm to fit the model.

The LASSO algorithm uses a modified loss function with a regularization term:

$$L(w) = ||Xw-y||_2^{2} + \alpha ||w'||_1$$

where w' is the vector composed of all entries of w exclusing the very last entry.

In LASSO algorithm, the entries of the weight vector w is very small, tending to force entries of the weight vector to be exactly zero.

#### Experiment on LASSO Algorithm

In this experiment, I will use similar method as used in Part II. I will increase the p_features all the way to n_train -1 to see how scores change along the way.

##### Experiment 1: alpha = 0.001

```python
from sklearn.linear_model import Lasso
L = Lasso(alpha = 0.001)

n_train = 100
n_val = 100
p_features = 1
noise = 0.2

L_training_score_history = []
L_validation_score_history = []
for i in range (n_train-1):
    X_train, y_train, X_val, y_val = LR_data(n_train, n_val, p_features, noise)
    L.fit(X_train, y_train)
    L_training_score_history.append(L.score(X_train, y_train).round(4))
    L_validation_score_history.append(L.score(X_val, y_val).round(4))
    p_features += 1
plt.plot(np.arange(n_train-1), L_training_score_history, label = "Lasso Training Score")
plt.plot(np.arange(n_train-1), L_validation_score_history, label = "Lasso Validation Score")
labels = plt.gca().set(xlabel = "p_features", ylabel = "Scores")
legend = plt.legend()
```
